### Import the module

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

### Read the current direction.

```python
data_path = r"/Users/linchunho/Developer/ATM-Chem/src/data/monthly_mean/"

os.chdir(data_path)
```

```python
for file in os.listdir() :
    if file.startswith(f"co2") :
    # if file.endswith(f"mlo.txt") : 
        file_path = f"{data_path}/{file}"
        data = pd.read_csv(file_path, skiprows=53, sep="\s+")
        data = data.replace(-0.99, 0)
        data = data.replace(-9.99, 0)
        data = data.replace(-1   , 0)
        data.columns = ["year", "month", "date", "avg", "de-s", "day", "std", "unc"]

        print(data)
```

```python
plt.figure(figsize=(20, 15))
plt.plot(data["date"], data["de-s"], "k-o")
plt.plot(data["date"], data["avg"], "r-o")

plt.xlabel("Year")
plt.ylabel("CO2 mole fraction (ppm)")
plt.title("Atomospheric CO2 at Mauna Loa Observatory")

plt.grid()
plt.savefig("../../imgs/co2_monthly_mean.jpg", dpi=300)
plt.show()
```

```python
for file in os.listdir() :
    if file.endswith(f"mlo.txt") :
        file_path = f"{data_path}/{file}"
        data = pd.read_csv(file_path, skiprows=771, sep="\s+")
        data = data.replace(-0.99, 0)
        data = data.replace(-9.99, 0)
        data = data.replace(-1   , 0)
        data.columns = ["year", "month", "date", "avg", "de-s", "day", "std", "unc"]

        print(data)
```

```python
plt.figure(figsize=(20, 15))
plt.plot(data["date"], data["de-s"], "k-o")
plt.plot(data["date"], data["avg"], "r-o")

plt.xlabel("Year")
plt.ylabel("CO2 mole fraction (ppm)")
plt.title("Recently monthly Mean CO2 at Mauna Loa Observatory")

plt.grid()
plt.savefig("../../imgs/co2_recent_monthly_mean.jpg", dpi=300)
plt.show()
```
